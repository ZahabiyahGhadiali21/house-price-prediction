```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# deciding upon which factor to take for prediction
```

```python
labels = data['price']

# Convert the 'date' column into a binary indicator for the year 2014
data['date'] = data['date'].apply(lambda x: 1 if x == 2014 else 0)

# Select relevant columns as features
selected_columns = ['bedrooms', 'bathrooms', 'sqft_living', 'floors', 'waterfront', 'view', 'condition', 'yr_built', 'yr_renovated']
selected_data = data[selected_columns]
```

```python
x_train, x_test, y_train, y_test = train_test_split(selected_data, labels, test_size=0.8, random_state=42)
```

```python
# Check for missing values in the selected_data DataFrame
missing_values = selected_data.isna().sum()

# Print the columns with missing values and their corresponding counts
print("Columns with Missing Values:")
print(missing_values[missing_values > 0])
```

```python
# There is no missing value so there is no issue.
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(x_train)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(x_test)
```

```python
# Best parameter choosing for Linear Regression and XGBoost:
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler


# Defining models and their parameter grids
models = {
    'Linear Regression': (LinearRegression(), {}),
    'XGBoost': (XGBRegressor(random_state=42), {'n_estimators': [100, 200, 300], 'learning_rate': [0.01, 0.1, 0.3], 'max_depth': [3, 5, 7]})
}

# Evaluating models using cross-validation and print performance metrics
for name, (model, param_grid) in models.items():
    print(f"Evaluating {name}:")
    
    # Perform hyperparameter tuning with GridSearchCV
    grid_search = GridSearchCV(estimator=model,
                               param_grid=param_grid,
                               cv=5,
                               scoring='neg_mean_squared_error',  # Use negative MSE as scoring metric
                               verbose=2)
    
    grid_search.fit(X_train_scaled, y_train)
    
    # Get the best model
    best_model = grid_search.best_estimator_
    print(best_model)
    # Evaluate the best model with cross-validation
    cv_mse = -cross_val_score(best_model, X_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
    cv_r2 = cross_val_score(best_model, X_train_scaled, y_train, cv=5, scoring='r2')
    
    print(f"Best Parameters: {grid_search.best_params_}")
    print(f"Average Cross-Validation MSE: {cv_mse.mean()}")
    print(f"Average Cross-Validation R^2 Score: {cv_r2.mean()}\n")
```

```python
"""Evaluating Linear Regression:

Best Parameters: {}
Average Cross-Validation MSE: 47717496740.80761
Average Cross-Validation R^2 Score: 0.5883324497234844

Evaluating XGBoost:

Best Parameters: {'learning_rate': 0.1, 'max_depth': 3, 'n_estimators': 100}
Average Cross-Validation MSE: 44169157242.59664
Average Cross-Validation R^2 Score: 0.61810519334976"""
```

```python
#Linear regression model:
```

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train_scaled, y_train)

# Evaluate Linear Regression model on test set
y_pred_test_linear = linear_reg_model.predict(X_test_scaled)
test_mse_linear = mean_squared_error(y_test, y_pred_test_linear)
test_r2_linear = r2_score(y_test, y_pred_test_linear)
print("Linear Regression Test MSE:", test_mse_linear)
print("Linear Regression Test R^2 Score:", test_r2_linear)
```

```python
# Plotting predictions vs actual values for Linear Regression
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_test_linear, color='blue', label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Ideal Predictions')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Linear Regression: Actual vs Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#For Gradient Boosting Regressor using best paarmeter:
```

```python

# Train XGBoost model
xgb_model = XGBRegressor(random_state=42, n_estimators=100, learning_rate=0.1, max_depth=3)
xgb_model.fit(X_train_scaled, y_train)

# Evaluate XGBoost model on test set
y_pred_test_xgb = xgb_model.predict(X_test_scaled)
test_mse_xgb = mean_squared_error(y_test, y_pred_test_xgb)
test_r2_xgb = r2_score(y_test, y_pred_test_xgb)
print("XGBoost Test MSE:", test_mse_xgb)
print("XGBoost Test R^2 Score:", test_r2_xgb)
```

```python
# Plotting predictions vs actual values for XGBoost
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_test_xgb, color='green', label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Ideal Predictions')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('XGBoost: Actual vs Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Predicted vs Actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test_linear, color='blue', label='Linear Regression')
plt.scatter(y_test, y_pred_test_xgb, color='red', label='XGBoost')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Predicted vs Actual Prices')
plt.legend()
plt.show()

# Residual plot
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_test_linear, y_test - y_pred_test_linear, color='blue', label='Linear Regression')
plt.scatter(y_pred_test_xgb, y_test - y_pred_test_xgb, color='red', label='XGBoost')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate evaluation metrics for Linear Regression
mse_linear = mean_squared_error(y_test, y_pred_test_linear)
rmse_linear = mean_squared_error(y_test, y_pred_test_linear, squared=False)  # Calculate RMSE
mae_linear = mean_absolute_error(y_test, y_pred_test_linear)
r2_linear = r2_score(y_test, y_pred_test_linear)

# Calculate evaluation metrics for XGBoost
mse_xgb = mean_squared_error(y_test, y_pred_test_xgb)
rmse_xgb = mean_squared_error(y_test, y_pred_test_xgb, squared=False)  # Calculate RMSE
mae_xgb = mean_absolute_error(y_test, y_pred_test_xgb)
r2_xgb = r2_score(y_test, y_pred_test_xgb)

# Print the evaluation metrics
print("Linear Regression:")
print("MSE:", mse_linear)
print("RMSE:", rmse_linear)
print("MAE:", mae_linear)
print("R^2 Score:", r2_linear)
print("\nXGBoost:")
print("MSE:", mse_xgb)
print("RMSE:", rmse_xgb)
print("MAE:", mae_xgb)
print("R^2 Score:", r2_xgb)
```

```python

# Calculate residuals for Linear Regression
residuals_linear = y_test - y_pred_test_linear

# Calculate residuals for XGBoost
residuals_xgb = y_test - y_pred_test_xgb

# Plotting residuals for Linear Regression
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_test_linear, residuals_linear, color='blue', alpha=0.5, label='Linear Regression')

# Plotting residuals for XGBoost
plt.scatter(y_pred_test_xgb, residuals_xgb, color='red', alpha=0.5, label='XGBoost')

# Horizontal line at y=0
plt.axhline(y=0, color='black', linestyle='--')

# Labels and title
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.title('Residual Plot: Linear Regression vs XGBoost')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Define bins for actual and predicted values
bins = np.linspace(min(y_test.min(), y_pred_test_linear.min()), max(y_test.max(), y_pred_test_linear.max()), num=5)

# Digitize actual and predicted values into bins
actual_bins = np.digitize(y_test, bins)
predicted_bins_linear = np.digitize(y_pred_test_linear, bins)
predicted_bins_xgb = np.digitize(y_pred_test_xgb, bins)

# Compute confusion matrices
cm_linear = confusion_matrix(actual_bins, predicted_bins_linear)
cm_xgb = confusion_matrix(actual_bins, predicted_bins_xgb)

# Plot confusion matrices
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Linear Regression
axs[0].imshow(cm_linear, cmap='Blues', interpolation='nearest')
axs[0].set_title('Confusion Matrix (Linear Regression)')
axs[0].set_xlabel('Predicted Bins')
axs[0].set_ylabel('Actual Bins')
plt.colorbar(axs[0].imshow(cm_linear, cmap='Blues', interpolation='nearest'), ax=axs[0], shrink=0.5)

# XGBoost
axs[1].imshow(cm_xgb, cmap='Blues', interpolation='nearest')
axs[1].set_title('Confusion Matrix (XGBoost)')
axs[1].set_xlabel('Predicted Bins')
axs[1].set_ylabel('Actual Bins')
plt.colorbar(axs[1].imshow(cm_xgb, cmap='Blues', interpolation='nearest'), ax=axs[1], shrink=0.5)

plt.tight_layout()
plt.show()
```
